```python
# 모듈 로딩

# 내장 붓꽃 데이터셋
from sklearn.datasets import load_iris
# 의사결정트리 모델
from sklearn.tree import DecisionTreeClassifier
# 데이터셋 분리 함수
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#(2) 붓꽃 데이터 셋 로딩
iris = load_iris()
# 딕셔너리 형태
print(type(iris))
iris
```

```python
#(2) 붓꽃 데이터 셋 로딩
# data: 4개의 피처(feature)들로 이루어진 numpy 배열
iris_data = iris.data
iris_data
```

```python
# target: 붓꽃 데이터셋의 레이블(결정 값) 데이터로 된 numpy배열
iris_label = iris.target
print('iris target값:', iris_label)
print('iris target명:', iris.target_names)
```

```python
# 붓꽃 데이터셋을 자세히 보기 위해 DataFrame으로 변환
iris_df = pd.DataFrame(data=iris_data,
columns=iris.feature_names)
iris_df
```

```python
# 레이블 컬럼 추가
iris_df['label'] = iris_label
print(iris_df)
iris_df.head(3)
```

```python
#(3) 학습데이터/ 테스트 데이터 분리
# 데이터셋의 20%를 테스트 데이터로 사용
X_train, X_test, y_train, y_test = train_test_split(
iris_data,
iris_label,
test_size=0.2,
random_state=11
)
```

```python
#(4) 학습데이터셋으로 학습(Train)
# DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)
# 학습 수행
dt_clf.fit(X_train, y_train)
```

```python
#(5) 테스트 데이터셋으로 예측(Predict)
# 학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행.
pred = dt_clf.predict(X_test)
pred
```

```python
#(6) 예측 정확도 평가(accuracy_score)
from sklearn.metrics import accuracy_score
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```

```python
# (3) 교차검증
# KFold 객체의 split() 호출하면 폴드 별 학습용, 검증용 테스트의 row 인덱스를 array로 반환
for train_index, test_index in kfold.split(features_smart):
 # kfold.split()으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
 X_train, X_test = features_smart[train_index], features_smart[test_index]
 # Use .iloc to access elements by position
 y_train, y_test = label_smart.iloc[train_index], label_smart.iloc[test_index]

 #학습 및 예측
 dt_clf.fit(X_train, y_train)
 pred = dt_clf.predict(X_test)
 n_iter += 1

 # 반복 시 마다 정확도 측정
 accuracy = np.round(accuracy_score(y_test,pred), 4)
 train_size = X_train.shape[0]
 test_size = X_test.shape[0]
 print('\\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
 .format(n_iter, accuracy, train_size, test_size))
 print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))

 cv_accuracy.append(accuracy)
```

```python
# (3) 교차검증
# KFold 객체의 split() 호출하면 폴드 별 학습용, 검증용 테스트의 row 인덱스를 array로 반환
for train_index, test_index in kfold.split(features_smart):
 # kfold.split()으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
 X_train, X_test = features_smart[train_index], features_smart[test_index]
 y_train, y_test = label_smart.iloc[train_index], label_smart.iloc[test_index]

 #학습 및 예측
 dt_clf.fit(X_train, y_train)
 pred = dt_clf.predict(X_test)
 n_iter += 1

 # 반복 시 마다 정확도 측정
 accuracy = np.round(accuracy_score(y_test,pred), 4)
 train_size = X_train.shape[0]
 test_size = X_test.shape[0]
 print('\\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
 .format(n_iter, accuracy, train_size, test_size))
 print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))

 cv_accuracy.append(accuracy)
```

```python
# (4) 정확도
#개별 iteration별 정확도를 합하여 평균 정확도 계산
print('\\n## 평균 검증 정확도:', np.mean(cv_accuracy))
```

성 비율을 동일하게 나눔.

```python
# (4) 학습용/검증용 데이터 분리
# StratifiedKFold의 split() 호출시 반드시 레이블 데이터 셋도 추가 입력 필요
for train_index, test_index in skfold.split(features, label):
 # split()으로 반환된 인덱스를 이용하여 학습용, 검증용 테스트 데이터 추출
 X_train, X_test = features[train_index], features[test_index]
 y_train, y_test = label[train_index], label[test_index]

 #학습 및 예측
 dt_clf.fit(X_train , y_train)
 pred = dt_clf.predict(X_test)
 # 반복 시 마다 정확도 측정
 n_iter += 1
 accuracy = np.round(accuracy_score(y_test,pred), 4)
 train_size = X_train.shape[0]
 test_size = X_test.shape[0]

 print('\\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
 .format(n_iter, accuracy, train_size, test_size))
 print('#{0} 검증 세트 인덱스:{1}'.format(n_iter,test_index))
 cv_accuracy.append(accuracy)
```

```python
#(5) 정확도
# 교차 검증별 정확도 및 평균 정확도 계산
print('\\n## 교차 검증별 정확도:', np.round(cv_accuracy, 4))
print('## 평균 검증 정확도:', np.mean(cv_accuracy))
```

# **cross_val_score**

```python
# (1) 라이브러리 호출
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris
import numpy as np
```

```python
# (2) 데이터셋 로딩
iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)
dt_clf
```

```python
# (3) 피처 데이터셋
# 피처 데이터셋
data = iris_data.data
# 레이블 데이터셋
label = iris_data.target
label
```

```python
#(4) 정확도, 교차검증 셋 3개
# 성능 지표는 정확도(accuracy) , 교차 검증 세트는 3개
scores = cross_val_score(dt_clf , data , label ,
scoring='accuracy',cv=3)
print('교차 검증별 정확도:',np.round(scores, 4))
print('평균 검증 정확도:', np.round(np.mean(scores), 4))
```
