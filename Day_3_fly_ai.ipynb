```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 컬럼명을 다루기 쉽게 영어로 변경
df.columns = [
    'CrimeCategory', 'SubCategory', '00:00-02:59', '03:00-05:59', '06:00-08:59',
    '09:00-11:59', '12:00-14:59', '15:00-17:59', '18:00-20:59', '21:00-23:59',
    'unknown','sun','mon','tue','wed','thu','fri','sat'
]

# 분석을 위한 관련 컬럼만 선택
time_intervals = ['21:00-23:59', '00:00-02:59', '03:00-05:59']
filtered_data = df[['CrimeCategory'] + time_intervals]

# 각 시간대별 상위 5개 범죄 찾기
top_crimes = {}
for interval in time_intervals:
    top_crimes[interval] = filtered_data[['CrimeCategory', interval]].sort_values(by=interval, ascending=False).head(5)

# 시각화를 위해 결과를 하나의 DataFrame으로 결합
combined_data = pd.concat(top_crimes, axis=0).reset_index(drop=True)

# seaborn을 사용하여 데이터 시각화
plt.figure(figsize=(12, 8))
sns.barplot(x='CrimeCategory', y='value', hue='variable', df=pd.melt(combined_data, id_vars=['CrimeCategory'], value_vars=time_intervals))
plt.title('Top 5 Crimes by Time Interval')
plt.xlabel('Crime Category')
plt.ylabel('Number of Incidents')
plt.legend(title='Time Interval')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 컬럼명을 다루기 쉽게 영어로 변경
df.columns = [
    'CrimeCategory', 'SubCategory', '00:00-02:59', '03:00-05:59', '06:00-08:59',
    '09:00-11:59', '12:00-14:59', '15:00-17:59', '18:00-20:59', '21:00-23:59',
    'unknown','sun','mon','tue','wed','thu','fri','sat'
]

# 분석을 위한 관련 컬럼만 선택
time_intervals = ['21:00-23:59', '00:00-02:59', '03:00-05:59']
filtered_data = df[['CrimeCategory'] + time_intervals]

# 각 시간대별 상위 5개 범죄 찾기
top_crimes = {}
for interval in time_intervals:
    top_crimes[interval] = filtered_data[['CrimeCategory', interval]].sort_values(by=interval, ascending=False).head(5)

# 시각화를 위해 결과를 하나의 DataFrame으로 결합
combined_data = pd.concat(top_crimes, axis=0).reset_index(drop=True)

# pd.melt 결과를 변수에 저장
melted_data = pd.melt(combined_data, id_vars=['CrimeCategory'], value_vars=time_intervals)

# seaborn을 사용하여 데이터 시각화, melted_data를 사용
plt.figure(figsize=(12, 8))
sns.barplot(x='CrimeCategory', y='value', hue='variable', data=melted_data) # 'data' argument에 melted_data를 전달
plt.title('심야 시간 Top 5 범죄')
plt.xlabel('범죄 대분류')
plt.ylabel('범죄 횟수')
plt.legend(title='시간 간격')
plt.xticks(rotation=45)
plt.show()
```
